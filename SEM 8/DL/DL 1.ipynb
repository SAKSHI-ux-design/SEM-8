```python
import tensorflow as tf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
test_size=0.2
learning_rate=0.01
epochs=100
batch =10
```

```python
#Exam madhe jar aaplyala dataset provide kela tar he lilhaych aahe nahi jar kel tar yachya khalchi line
#but ig te aaplyala dataset provide kartil
#ani from sklearn.load_datset varun boston cha data nako load karu karan tyani tithun to dataset remove kela aahe error yeto
boston_data=pd.read_csv("1_boston_housing.csv")
#from sklearn.datasets import load_boston
boston_data
```

```python
boston_data = pd.read_csv("https://raw.githubusercontent.com/PROxZIMA/Academic-Codes/master/Semester%206/DSBDAL/A4/boston.csv")
boston_data
```

```python
boston_data.isna().sum()
```

```python
X=boston_data.drop('MEDV',axis=1)
y=boston_data.MEDV
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=test_size)
```

```python
#optional aahe zhal tarch kar
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms.fit(X_train)
X_train = mms.transform(X_train)
X_test = mms.transform(X_test)
```

```python
n_features=X.shape[1]
```

```python
model = Sequential()
model.add(Dense(10,activation='relu',input_shape=(n_features,)))
model.add(Dense(64,activation='sigmoid',))
model.add(Dense(1,activation='linear'))
model.compile(optimizer="adam",loss="mse",metrics=["mae"])
model.fit(X_train,y_train,epochs=epochs,batch_size=batch,verbose=1)
```

```python
mse_nn,mae_nn=model.evaluate(X_test,y_test)
```
